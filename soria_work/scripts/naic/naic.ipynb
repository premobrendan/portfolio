```python
import pandas as pd
import numpy as np

types = ["2023", "2024", "24Q1", "24Q2", "24Q3", "25Q1", "25Q2", "25Q3"]
endings = ["000", "001", "002", "003", "014", "015", "016", "097", "098"]
base = "data/X{}{}.csv"

dfs = []

for type in types:
    for ending in endings:
        # get file name of files
        filename = base.format(type, ending)
        if filename not in data:
            continue

        # read file with headers
        cols = headers[ending]
        df = pd.read_csv(filename, names=cols)

        # add date code and relevant metadata columns to df
        df["DATE_CODE"] = type
        if "Q" in type:
            year = int(type[:2])+2000
            quarter = int(type[3])
        else:
            year = int(type[:4])
            quarter = 0

        df["YEAR"] = year
        df["QUARTER"] = quarter
        df = df[["DATE_CODE", "YEAR", "QUARTER"]+cols]

        # get new file name based on ending and save
        new_filename = type + "_" + new_endings[ending] + ".csv"
        df.to_csv("clean_naic_data/"+new_filename, index=False)

```
