```python
import numpy as np

def check(L, M):
    for i in range(len(L)):
        if L[i] == M[i]:
            return i
    return -1

def run (pins):
    sim_num = 50000
    sim = 0
    count_list = np.zeros(len(pins))
    sim_data = np.zeros(len(pins))

    while sim < sim_num:
        sim += 1
        loop_break = True
        while loop_break:
            roll = np.random.randint(6) + np.random.randint(6)
            count_list[roll] = count_list[roll] + 1
            val = check(count_list, pins)
            if val != -1:
                sim_data[val] = sim_data[val] + 1
                count_list = np.zeros(len(pins))
                loop_break = False

    print("We have the following outcome for pin set", pins, ":")
    var_data = [0,0,0,0,0,0]
    for i in range(len(var_data)):
        var_data[i] = sim_data[i] / sim_num
    print("The percent chance for each horse number is the following:")
    print("Horses 2 & 12 : "+str(int(np.round(var_data[0]*100)))+"%")
    print("Horses 3 & 11 : "+str(int(np.round(var_data[1]*100)))+"%")
    print("Horses 4 & 10 : "+str(int(np.round(var_data[2]*100)))+"%")
    print("Horses 5 & 9  : "+str(int(np.round(var_data[3]*100)))+"%")
    print("Horses 6 & 8  : "+str(int(np.round(var_data[4]*100)))+"%")
    print("Horse  7      : "+str(int(np.round(var_data[5]*100)))+"%")
    """
    for i in range(len(sim_data)):
        print("Percent chance for the Horse whose roll is", i+2, "to win are: "+str(int(np.round(sim_data[i]*100/sim_num)))+"%")
    """


print("Original pin set:")
pins = np.asarray([3,6,8,11,14,16,14,11,8,6,3])
run(pins)
print()
print("Inner Adapted pin set:")
pins = np.asarray([3,6,8,11,13,14,13,11,8,6,3])
run(pins)
print()
print("*Remark*: from this, we can see that the issue is less so that the inner pieces aren't winning enough,")
print("but the outer is winning MORE than it should, so we increase their pin number by 1")
print()
print("Outer Adapted pin set:")
print("*Remark*: a good place to make edits for optimization")
pins = np.asarray([4,6,8,11,14,16,14,11,8,6,4])
run(pins)
print()
print("Advanced Outer Adapted pin set:")
print("*Remark*: a good step forward, but we can do better")
pins = np.asarray([4,7,9,11,14,16,14,11,9,7,4])
run(pins)
print()
print("Advanced Outer + Inner Adapted pin set:")
print("*Remark*: pretty good with a slight skew for the center pieces")
pins = np.asarray([4,7,9,11,13,15,13,11,9,7,4])
run(pins)
print()
print("Outer + Advanced Inner Adapted pin set:")
print("*Remark*: might be the best one without the game longer due to higher pin numbers, but I make no claim this is optimal")
pins = np.asarray([4,6,8,10,12,14,12,10,8,6,4])
run(pins)
```
